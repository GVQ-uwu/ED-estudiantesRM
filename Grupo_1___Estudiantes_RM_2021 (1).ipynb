# Análisis de Relaciones entre Variables en Matrículas de Educación Superior

Objetivos:
Analizar relaciones entre variables cuantitativas y cualitativas en datos de matrículas de educación superior, utilizando técnicas de agrupación, medidas centrales y gráficos.

Contexto:
Utilizaremos la base de datos de matrículas en educación superior de la Región Metropolitana 2021. Las preguntas a resolver son:
1. ¿Existe relación entre la duración de la carrera (cuantitativa discreta) y el tipo de carrera (cualitativa nominal)?
2. ¿Cómo varía la preferencia por la jornada académica (cualitativa nominal) según la edad de los estudiantes (cuantitativa continua)?

```python
# ------------- Pregunta 1: Relación entre Duración y Carrera -------------
```

```python
# Para este estudio trabajaremos con 3 carreras de tegnología
df_pregunta1 = df_pregunta1[
    (df_pregunta1['NOMBRE CARRERA'] == "ADMINISTRACION DE REDES Y TELECOMUNICACIONES") |
    (df_pregunta1['NOMBRE CARRERA'] == "INGENIERIA EN INFORMATICA") |
    (df_pregunta1['NOMBRE CARRERA'] == "ANALISTA PROGRAMADOR")
]
display(df_pregunta1.head())
```

```python
# Muestra de las Tablas Filtradas
muestra = df_pregunta1.shape[0]
print(f"La Muestra de las 3 carreras seleccionadas es de {muestra} Estudiantes de Universidades en la Region Metropolitana")
```

```python
# Frecuencias de Duracion Semestres DURACION TOTAL CARRERA (SEMESTRES)
frecu_semestre = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].value_counts()
tabla_semestres = pd.DataFrame({
    "Frecuencia Absoluta": frecu_semestre,
    "Frecuencia Relativa (%)": frecu_semestre / muestra * 100,
    "Frecuencia Absoluta Acumulada": frecu_semestre.cumsum(),
    "Frecuencia Relativa Acumulada (%)": (frecu_semestre / muestra * 100).cumsum()
})
round(tabla_semestres,1)
```

```python
# Grafico circular para visualizar la distribución de la cantidad de semestres.
import matplotlib.pyplot as plt

nombre_carrera = tabla_semestres.index
porcen_carrera = tabla_semestres['Frecuencia Relativa (%)']
fig, ax = plt.subplots(figsize = (6, 6))
circular_ocup = ax.pie(porcen_carrera, labels = nombre_carrera, autopct = '%.1f%%')
ax.set_title('Porcentajes de Alumnos Inscritos entre las 3 Carreras')
plt.show()
```

```python
# Gráfico de barras para visualizar la distribución de frecuencia de la duración total de la carrera en semestres para las carreras seleccionadas.

cant_semestre = tabla_semestres.index
frecu_semestre = tabla_semestres['Frecuencia Absoluta']
fgi, ax = plt.subplots(figsize = (6, 6))
grafico_leng = ax.bar(cant_semestre, frecu_semestre, color = 'violet', edgecolor = 'black')
ax.set_title('Frecuencia de duracion en Semestres de las 3 Carreras Seleccionadas')
ax.set_xlabel('Cantidad de Semestres')
ax.set_ylabel('Frecuencia de duracion de Semestres')
ax.bar_label(grafico_leng)
plt.xticks(rotation = 15)
plt.show()
print()
```

```python
# Identificar la mayor frecuencia
mayor_frecuencia = tabla_semestres['Frecuencia Absoluta'].max()
duracion_mayor_frecuencia = tabla_semestres['Frecuencia Absoluta'].idxmax()

print(f"La mayor frecuencia de estudiantes se encuentra en carreras con una duración de {duracion_mayor_frecuencia} semestres, con un total de {mayor_frecuencia} estudiantes.")
```

```python
# Identificar la menor frecuencia
menor_frecuencia = tabla_semestres['Frecuencia Absoluta'].min()
duracion_menor_frecuencia = tabla_semestres['Frecuencia Absoluta'].idxmin()

print(f"La menor frecuencia de estudiantes se encuentra en carreras con una duración de {duracion_menor_frecuencia} semestres, con un total de {menor_frecuencia} estudiantes.")
```

```python
# Calcular medidas de tendencia central de la duracion de las carreras
mean_semestre = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].mean()
median_semestre = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].median()
moda_semestre = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].mode()

print(f"La duración promedio de las carreras seleccionadas es de {mean_semestre:.1f} semestres")
print(f"El 50% de las carreras seleccionadas tienen una duración de a lo mas de {median_semestre:.1f} semestres")
print(f"La duración más frecuente de las carreras seleccionadas es de {moda_semestre[0]} semestres.")
```

```python
# Calcular medidas no centrales (percentiles)

# Calculamos algunos percentiles clave para entender mejor la distribución
percentiles = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].quantile([0.25, 0.50, 0.75])

print("Percentiles de la duración total de la carrera (en semestres):")
print('\n')

display(percentiles)

# Interpretación de algunos percentiles:
print('\n')
print(f"El 25% de las carreras tienen una duración de a lo más {percentiles[0.25]:.1f} semestres.")
print(f"El 50% de las carreras tienen una duración de a lo más {percentiles[0.5]:.1f} semestres.")
print(f"El 75% de las carreras tienen una duración de a lo más {percentiles[0.75]:.1f} semestres.")
```

```python
# Determinar el rango de distribución
rango = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].max() - df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].min()
varianza = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].var()
desviacion = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].std()
q25 = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].quantile(0.25)
q75 = df_pregunta1['DURACION TOTAL CARRERA (SEMESTRES)'].quantile(0.75)

print(f"El rango de duración es {rango} semestres")
print(f"La varianza de la duración es {varianza:.2f}")
print(f"La desviación estándar de la duración es {desviacion:.2f}")
print(f"El 25% de las carreras dura hasta {q25} semestres y el 75% hasta {q75} semestres")
```

```python
# Gráfico bivariado
import matplotlib.pyplot as plt

df_pregunta1 = df_pregunta1.groupby(['NOMBRE CARRERA', 'DURACION TOTAL CARRERA (SEMESTRES)'],observed=True ).size().unstack(fill_value=0)
df_pregunta1

fig, ax = plt.subplots(figsize=(10, 6))

df_pregunta1.plot(kind='bar', ax=ax)
for container in ax.containers:
    ax.bar_label(container, fmt='%d')

ax.set_title('Distribución de Semestres entre las 3 Carreras Elegidas')
ax.set_xlabel('Carreras')
ax.set_ylabel('Cantidad de Semestres')
ax.tick_params(axis='x', rotation=0)
ax.legend(title='Cantidad de Semestres por Color')

plt.tight_layout()
plt.show()
```
